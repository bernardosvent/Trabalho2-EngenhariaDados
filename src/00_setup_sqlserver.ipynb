# Passo 0: Setup do SQL Server
Este notebook gera dados fictícios e insere no SQL Server usando PySpark via JDBC.

```python
from pyspark.sql import SparkSession

# Inicializa o Spark com o driver do SQL Server
spark = SparkSession.builder \
    .appName("Setup-SQLServer") \
    .config("spark.jars.packages", "com.microsoft.sqlserver:mssql-jdbc:12.6.1.jre11") \
    .getOrCreate()

# Configurações de conexão
jdbc_url = "jdbc:sqlserver://127.0.0.1:1433;databaseName=master;encrypt=false;trustServerCertificate=true"
jdbc_props = {
    "user": "sa",
    "password": "SqlServer@2024!",
    "driver": "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}

# Dados iniciais de uma partida (Simulação: Flamengo x Vasco)
data = [
    (1, 101, 10, "Gol", 15, False),             # Gol do camisa 10
    (2, 101, 5, "Cartão Amarelo", 30, False),    # Falta do volante
    (3, 101, 9, "Gol", 44, False)              # Gol do centroavante
]

columns = ["id_evento", "id_partida", "id_jogador", "tipo_evento", "minuto_jogo", "revisado_var"]

df_eventos = spark.createDataFrame(data, columns)

# Escreve no SQL Server
df_eventos.write.jdbc(url=jdbc_url, table="fato_evento", mode="overwrite", properties=jdbc_props)
print("Tabela 'fato_evento' criada com sucesso no SQL Server!")
```
