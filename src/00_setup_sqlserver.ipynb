# Passo 0: Setup do SQL Server
Este notebook gera dados fictícios e insere no SQL Server usando PySpark via JDBC.

```
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType

# Inicializa o Spark com o driver do SQL Server
spark = SparkSession.builder \
    .appName("Setup-SQLServer") \
    .config("spark.jars.packages", "com.microsoft.sqlserver:mssql-jdbc:12.6.1.jre11") \
    .getOrCreate()

# Configurações de conexão
jdbc_url = "jdbc:sqlserver://localhost:1433;databaseName=master;encrypt=false;trustServerCertificate=true"
jdbc_props = {
    "user": "sa",
    "password": "SqlServer@2024!",
    "driver": "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}

# Dados de exemplo: Clientes
clientes_data = [
    (1, "Ana Silva", "ana@email.com", 28),
    (2, "Carlos Souza", "carlos@email.com", 35),
    (3, "Maria Oliveira", "maria@email.com", 42)
]
schema_clientes = StructType([
    StructField("id", IntegerType(), False),
    StructField("nome", StringType(), True),
    StructField("email", StringType(), True),
    StructField("idade", IntegerType(), True)
])

df_clientes = spark.createDataFrame(clientes_data, schema=schema_clientes)

# Escreve no SQL Server
df_clientes.write.jdbc(url=jdbc_url, table="clientes", mode="overwrite", properties=jdbc_props)
print("Tabela 'clientes' criada com sucesso no SQL Server!")
```
